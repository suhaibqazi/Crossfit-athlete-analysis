### Start Classification - changing one tier at a time

```python
Tier_4_notes = ['Always On','Launch']
Tier_5_exclude = ['Tarriff','New Product']
```

```python
#Tier 6

def tier6_rules(raw_Df):
    val=''
    #print (raw_Df['Tier 4'])
    if (str(raw_Df['Tier 5']) == 'Android') & (str(raw_Df['Tier 6']) == 'Pay Monthly Handset'):
        val = 'Pay Monthly Android'
    elif (str(raw_Df['Tier 5']) == 'iPhone') & (str(raw_Df['Tier 6']) == 'Pay Monthly Handset'):
        val = 'Pay Monthly iPhone'
    elif (str(raw_Df['Tier 5']) == 'Other Handset') & (str(raw_Df['Tier 6']) == 'Pay Monthly Handset'):
        val = 'Pay Monthly Handset'
    elif (str(raw_Df['Tier 5']) == 'Tablet') & ((str(raw_Df['Tier 6']) == 'MBB') or (str(raw_Df['Tier 6']) == 'Pay Monthly Tablet')):
        val = 'PM Tablet - MBB'
       #prop led 
    elif (str(raw_Df['Tier 5']) == 'Prop Led Android') & (str(raw_Df['Tier 6']) == 'Pay Monthly Handset'):
        val = 'Pay Monthly Android'
    elif (str(raw_Df['Tier 5']) == 'Prop Led iPhone') & (str(raw_Df['Tier 6']) == 'Pay Monthly Handset'):
        val = 'Pay Monthly iPhone'
    elif (str(raw_Df['Tier 5']) == 'O2 Search'):
        val = 'O2 Brand Search'
    else: val = raw_Df['Tier 6']

    return val 

#Tier 5 

def tier5_rules(raw_Df):
    val=''
    if any(word in str(raw_Df['Tier 4']) for word in Tier_4_notes):
        val=''
    elif (str(raw_Df['Tier 5']) == 'O2 Search'):
        val = ''
    else: val = raw_Df['Tier 5']   
   
    return val

#Tier 4 
def tier4_rules(raw_Df):
    val=''
    if (str(raw_Df['Tier 5']) == 'O2 Search'):
        val = 'Always On'
    else: val = raw_Df['Tier 4']   
   
    return val

#Tier 1
def tier1_rules(raw_Df):
    val=''
    if (str(raw_Df['Tier 5']) == 'O2 Search'):
        val = 'Trading'
    else: val = raw_Df['Tier 1']   
   
    return val
```

```python
df_unique2.loc[:,'Tier 6']=df_unique.apply(tier6_rules,axis=1)
df_unique2.loc[:,'Tier 4']=df_unique.apply(tier4_rules,axis=1)
df_unique2.loc[:,'Tier 1']=df_unique.apply(tier1_rules,axis=1)
df_unique2.loc[:,'Tier 5']=df_unique.apply(tier5_rules,axis=1)
```

```python
R_and_F = ['r&f','reach','randf']
```

```python
#R and F rules

def randf_rules(raw_Df):
    val='' 
    if any(word in str(raw_Df['Campaign']).lower() for word in R_and_F):
        val='Paid Social Upper'
    elif (raw_Df['Driver'] == 'Paid Social'):
        val = 'Paid Social Lower'
    else: val = raw_Df['Driver']
    
    return val

```
