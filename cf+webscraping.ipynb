```python
stats_tables = soup.find('table', class_='stats')
stats_tables_All = soup.findAll('table', class_='stats')
```

```python
#Generate lists
A=[]
B=[]

for el in stats_tables_All:
    for row in el.find_all("tr"):
      
        cells = row.find_all('td')
        workout=row.findAll('th')

        A.append(cells[0].find(text=True))
        B.append(workout[0].find(text=True))        
    
```
